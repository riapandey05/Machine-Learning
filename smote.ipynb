Smote is a technique used in ml to address imbalanced datasets. Smote involved generating synthetic instances of the minority class by interpolating between existing instances

```python
# interpolating means we are creating ou own symthetic data.
# smote tries to join two nearest points and create a new point between them.
# now data created also has variance unlike oversampling. in which there in no variance
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0, n_features=2, n_clusters_per_class=1,weights=[0.9, 0.1], random_state=42)
```

```python
import pandas as pd
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
oversample = SMOTE()
```

```python
X,y = oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
